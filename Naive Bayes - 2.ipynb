```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv("D:\PYTHON\Python For Machine Learning\Data\\Wine.csv")
df.head(20)
```

```python
df.columns = ['Label','Alcohol',
'Malic acid',
'Ash',
'Alcalinity of ash',
'Magnesium',
'Total phenols',
'Flavanoids',
'Nonflavanoid phenols',
'Proanthocyanins',
'Color intensity',
'Hue',
'OD280/OD315 of diluted wines',
'Proline']
```

# Splitting the data into x and y where x contains all the columns and y contains the target values

```python
x, y = df.columns[1:], df.columns[0]
```

# Splitting the dataset into training and Testing set

```python
xtrain, xtest, ytrain, ytest = train_test_split(df[x],df[y], test_size=.2, random_state=7)
```

```python
clf = MultinomialNB()
```

```python
targets = ytrain.values
```

# Fitting the training set values

```python
clf.fit(xtrain, ytrain)
```

```python
y_pred = clf.predict(xtest)
```

# Calculating the score for our trained Model

```python
clf.score(xtest, ytest)
```
